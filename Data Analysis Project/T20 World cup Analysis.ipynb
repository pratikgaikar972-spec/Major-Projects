# Here we are importing the required libraries for data analysis and visualization.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Here we are loading the ICC Men’s T20 World Cup dataset into a Pandas DataFrame
# and displaying the first five rows to understand the structure of the data.

```python
world_cup=pd.read_csv("ICC Mens T20 Worldcup.csv")
world_cup.head()
```

# Here we are checking the structure of the dataset, including column names, data types, and non-null values.

```python
print(world_cup.info())
```

```python
print(world_cup.isnull().sum())# checking null values
```

```python
print(world_cup["1st Team"].unique())#exploring all the teams 
print(world_cup["2nd Team"].unique())
```

```python
print(world_cup[world_cup["Stage"]=="Final"]["Winners"])#winner of the World Cup
```

# Teams with the highest count are the most consistent match winners and are top performers in the tournament

```python
print(world_cup["Winners"].value_counts())
```

# Here we calculate the number of matches won by each team and visualize the results using a bar chart.

```python
winners = world_cup["Winners"].value_counts().reset_index(name="wins")
plt.figure(figsize=(6,3))
sns.barplot(winners,x="Winners",y="wins",color="#00B7B5",edgecolor="k")
plt.title("Most Successful Teams in T20 World Cup",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.xlabel("Teams",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.ylabel("Matches Won",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.xticks(rotation=90)
plt.show()
```

# here we have calulated toss_winner as winner of the match

```python
toss_win = world_cup["Toss Winning"].value_counts().reset_index(name="wins")
plt.figure(figsize=(6,3))
sns.barplot(toss_win,x="Toss Winning",y="wins",palette="bright",edgecolor="k")
plt.title("Toss Winners Count",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.ylabel("Frequency",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.xticks(rotation=90)
plt.show()
```

# here we have calculated top 10 venues where most matches are played

```python
venue = world_cup["Venue"].value_counts().reset_index(name="counts").head(10)
plt.figure(figsize=(5,3))
sns.barplot(venue,x="counts",y="Venue",orient="h",color="#3B9797",edgecolor="k")
plt.title("Top 10 Most Used Venues",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.xlabel("counts",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.ylabel("Venue",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.show()
```

# here we have calculated types of match wins 

```python
win_type=world_cup["Won by"].value_counts()
plt.pie(win_type,labels=["Runs","Wickets","Rain"],colors=["#73C8D2","#FFA4A4","#F5F1DC"],shadow=True,wedgeprops={"edgecolor":"k","linewidth":1.3})
plt.title("Match Win Type Distribution",fontdict={"color":"#11224E","fontfamily":"monospace"})
plt.show()
```

# here we have calculatd matches won by each team

```python
win_per=world_cup.groupby("Winners").size().reset_index(name="wins")
winners = win_per["wins"].sum()
win_per["win_percentage"]=(win_per["wins"]/winners)*100
win_per=win_per.sort_values(by="win_percentage",ascending=False)
sns.barplot(win_per,x="Winners",y="win_percentage",hue="wins",palette="muted",edgecolor="k",saturation=0.9)
plt.xticks(rotation=90)
plt.show()
```

# Batsman with highest Score 

```python
Top_scorer=world_cup[["Highest Score","Top Scorer"]].sort_values(by="Highest Score",ascending=False).head(15)
Top_scorer.drop([5,32,29,22],inplace=True)
sns.barplot(Top_scorer,y="Highest Score",x="Top Scorer",orient="y",gap=2,errorbar=("ci",0),dodge="auto",hue=(),edgecolor="black",linewidth=1,palette="bright")
Top_scorer["Highest Score"]=Top_scorer["Highest Score"].astype(int)
plt.xticks(rotation=90)
plt.show()
```
