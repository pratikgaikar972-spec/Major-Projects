```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("mymoviedb.csv",lineterminator = "\n")
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

Exploration Summary
*we have a dataframe consissting of 9827 rows and 9 columns.\n
*our dataset looks a bit tidy with no Nans nor duplicate values.\n
*Release_Date column needs to be casted into date time and to extract only the year value.\n
*Overview and Poster url wouldn't be so useful during analysis so we'll drop them.\n
*Vote_Average better be categorized for proper analysis.\n
*Genre column has comma seperated values and white spaces that needs to be handled and casted into category.\n

```python
data.head()
```

```python
data["Release_Date"]= pd.to_datetime(data["Release_Date"])
print(data["Release_Date"].dtypes)
```

```python
data["Release_Date"]=data["Release_Date"].dt.year
data["Release_Date"].dtypes
```

```python
data.head(2)
```

Dropping Columns

```python
data.drop(["Overview","Poster_Url"],axis=1,inplace=True)
data.head()
```

```python
data.loc[data["Vote_Average"] < 4, "Vote_Category"] = "not_popular"
data.loc[(data["Vote_Average"] >= 4) & (data["Vote_Average"] < 6), "Vote_Category"] = "below_avg"
data.loc[(data["Vote_Average"] >= 6) & (data["Vote_Average"] < 7.5), "Vote_Category"] = "average"
data.loc[data["Vote_Average"] >= 7.5, "Vote_Category"] = "popular"
```

```python
data["Vote_Category"].fillna("not_popular")
data["Vote_Category"].unique()
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data["Genre"]=data["Genre"].str.split(", ")

data = data.explode("Genre").reset_index(drop=True)
data.head()
```

```python
data.nunique()
```

# Data Visualization

```python
sns.set_style("whitegrid")
```

# what is the most frequent genre of movies realeased on Netflix

```python
sns.catplot(data,y="Genre",kind="count",order=data["Genre"].value_counts().index)
plt.title("Genre Distribution")
plt.show()
```

# Which Category has highest votes

```python
sns.countplot(
    data=data,
    y="Genre",
    hue="Vote_Category",
    palette="dark"
)
plt.title("Genre vs Vote Category Count")
plt.show()
```
